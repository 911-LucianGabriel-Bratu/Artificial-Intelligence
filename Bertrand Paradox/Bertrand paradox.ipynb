```python
#Bertrand paradox case 1
import numpy as np
import matplotlib.pyplot as plt
import math
import random

theta = np.linspace(0, 2*np.pi, 150)

radius = 1

x1 = radius*np.cos(theta) 
y1 = radius*np.sin(theta) 

figure, axes = plt.subplots( 1 ) 

axes.plot(x1, y1)

axes.set_aspect( 1 )

x = np.array([0,120,240,0])
y = np.array([0,120,240,0])
x = np.cos( x * np.pi / 180 )
y = np.sin( y * np.pi / 180 )

plt.plot(x, y, color='red' )

x_coords = []
y_coords = []

for i in range(0,25):
    x_coord = random.random()
    y_coord = 1 - x_coord**2
    y_coord = math.sqrt(y_coord)
    x_coords.append(x_coord)
    y_coords.append(y_coord)
    
    x_coord = random.random()
    x_coord = -x_coord
    y_coord = 1 - x_coord**2
    y_coord = math.sqrt(y_coord)
    x_coords.append(x_coord)
    y_coords.append(y_coord)
    
    x_coord = random.random()
    y_coord = 1 - x_coord**2
    y_coord = math.sqrt(y_coord)
    y_coord = -y_coord
    x_coords.append(x_coord)
    y_coords.append(y_coord)
    
    x_coord = random.random()
    y_coord = 1 - x_coord**2
    y_coord = math.sqrt(y_coord)
    x_coord = -x_coord
    y_coord = -y_coord
    x_coords.append(x_coord)
    y_coords.append(y_coord)

plt.plot(x_coords, y_coords ,color = "yellow")

valid_lengths = 0
triangle_side = math.sqrt(3)

for i in range(0,99):
    distance = math.sqrt((x_coords[i+1]+x_coords[i])**2 + (y_coords[i+1]+y_coords[i])**2)
    if(distance > triangle_side):
        valid_lengths += 1
    
plt.show()

print("Probability in case 1 is:" + str(valid_lengths/100))
```

```python
#Bertrand paradox case 2
import numpy as np
import matplotlib.pyplot as plt
import math
import random

theta = np.linspace(0, 2*np.pi, 150)

radius = 1

x1 = radius*np.cos(theta) 
y1 = radius*np.sin(theta) 

figure, axes = plt.subplots( 1 ) 

axes.plot(x1, y1)

axes.set_aspect( 1 )

x = np.array([0,120,240,0])
y = np.array([0,120,240,0])
x = np.cos( x * np.pi / 180 )
y = np.sin( y * np.pi / 180 )

plt.plot(x, y, color='red')

x_coords = [0]
y_coords = [0]

x_coord = random.random()
y_coord = 1 - x_coord**2
y_coord = math.sqrt(y_coord)
x_coords.append(x_coord)
y_coords.append(y_coord)

plt.plot(x_coords, y_coords, color='black')

ratio = x_coord/100

new_xs = []
new_ys = []

new_x_coord = ratio

print("Ratio=" + str(ratio))

for i in range(0,100):
    new_y_coord = (new_x_coord*y_coord)/x_coord
    new_xs.append(new_x_coord)
    new_ys.append(new_y_coord)
    new_x_coord+=ratio
    
radius_slope = y_coord/x_coord
perpendicular_slope = -1/radius_slope

circle_points_xs = []
circle_points_ys = []

for i in range(0,100):
    a = perpendicular_slope**2 + 1
    b = (2*perpendicular_slope*new_ys[i]) - (2*(perpendicular_slope**2)*new_xs[i])
    c = ((new_xs[i]**2)*(perpendicular_slope**2)) - (2*perpendicular_slope*new_xs[i]*new_ys[i]) + (new_ys[i]**2) - 1
    delta = b**2 - 4*a*c
    
    if delta > 0:
        x_circle_point_1 = (-b + math.sqrt(delta))/(2*a)
        x_circle_point_2 = (-b - math.sqrt(delta))/(2*a)

        y_circle_point_1 = math.sqrt(1-x_circle_point_1**2)
        y_circle_point_2 = math.sqrt(1-x_circle_point_2**2)

        circle_points_xs.append(x_circle_point_1)
        circle_points_xs.append(x_circle_point_2)
        circle_points_ys.append(y_circle_point_1)
        circle_points_ys.append(y_circle_point_2)
        
        plt.plot([x_circle_point_1, x_circle_point_2], [y_circle_point_1, y_circle_point_2], color = "blue")

plt.plot(new_xs, new_ys, color='orange')

valid_lengths = 0
triangle_side = math.sqrt(3)

for i in range(0,50):
    distance = math.sqrt((circle_points_xs[2*i+1] - circle_points_xs[2*i])**2 + (circle_points_ys[2*i+1] - circle_points_ys[2*i])**2)
    if(distance > triangle_side):
        valid_lengths += 1
    
plt.show()

print("Probability in case 2 is:" + str(valid_lengths/100))
```

```python
#Bertrand paradox case 3
import numpy as np
import matplotlib.pyplot as plt
import math
import random

theta = np.linspace(0, 2*np.pi, 150)

radius = 1

x1 = radius*np.cos(theta) 
y1 = radius*np.sin(theta) 

figure, axes = plt.subplots( 1 ) 

axes.plot(x1, y1)

axes.set_aspect( 1 )

x2 = 1/2*np.cos(theta)
y2 = 1/2*np.sin(theta)

axes.plot(x2, y2)

x = np.array([0,120,240,0])
y = np.array([0,120,240,0])
x = np.cos( x * np.pi / 180 )
y = np.sin( y * np.pi / 180 )

plt.plot(x, y, color='red')

x_coords = []
y_coords = []

for i in range(0, 25):
    x_coord = random.random()
    y_coord = random.random()
    while(x_coord**2 + y_coord**2 >= 1):
        x_coord = random.random()
        y_coord = random.random()
    x_coords.append(x_coord)
    y_coords.append(y_coord)
    x_coords.append(-x_coord)
    y_coords.append(y_coord)
    x_coords.append(x_coord)
    y_coords.append(-y_coord)
    x_coords.append(-x_coord)
    y_coords.append(-y_coord)

for i in range(0, 100):
    OM_slope = y_coords[i]/x_coords[i]
    perpendicular_slope = -1/OM_slope
    a = perpendicular_slope**2 + 1
    b = (2*perpendicular_slope*y_coords[i]) - (2*(perpendicular_slope**2)*x_coords[i])
    c = ((x_coords[i]**2)*(perpendicular_slope**2)) - (2*perpendicular_slope*x_coords[i]*y_coords[i]) + (y_coords[i]**2) - 1
    delta = b**2 - 4*a*c
    
    if delta > 0:
        x_circle_point_1 = (-b + math.sqrt(delta))/(2*a)
        x_circle_point_2 = (-b - math.sqrt(delta))/(2*a)

        y_circle_point_1 = math.sqrt(1-x_circle_point_1**2)
        y_circle_point_2 = math.sqrt(1-x_circle_point_2**2)
        
        plt.plot([x_circle_point_1, x_circle_point_2], [y_circle_point_1, y_circle_point_2], color='purple')

plt.plot(x_coords, y_coords, 'o')
    
valid_points = 0
for i in range(0, 100):
    point_value = math.sqrt(x_coords[i]**2 + y_coords[i]**2)
    if point_value < 1/2:
        valid_points += 1
    
plt.show()

print("Probability in case 3 is:" + str(valid_points/100))
```
