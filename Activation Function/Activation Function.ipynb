```python
#1.Compute the gradient for the sigmoid activation function in 2 points using pytorch and check it with the known explicit formula

import torch

# Define the input points
x1 = torch.tensor([1.0], requires_grad=True)
x2 = torch.tensor([0.0], requires_grad=True)

# Compute the sigmoid function at the input points
y1 = torch.sigmoid(x1)
y2 = torch.sigmoid(x2)

# Compute the gradients using PyTorch
y1.backward()
grad1 = x1.grad

y2.backward()
grad2 = x2.grad

# Compare with the known explicit formula for the sigmoid gradient
exp_grad1 = y1 * (1 - y1)
exp_grad2 = y2 * (1 - y2)

# Print the results
print("Gradient at x1 = 1.0 computed by PyTorch:", grad1.item())
print("Gradient at x1 = 1.0 using the explicit formula:", exp_grad1.item())
print("Gradient at x2 = 0.0 computed by PyTorch:", grad2.item())
print("Gradient at x2 = 0.0 using the explicit formula:", exp_grad2.item())
```

```python
#4.Write in python a function to plot the sigmoid activation function and its gradient using matplotlib

import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_gradient(x):
    return sigmoid(x) * (1 - sigmoid(x))

def plot_sigmoid():
    x = np.linspace(-10, 10, 1000)
    y = sigmoid(x)
    dy = sigmoid_gradient(x)

    plt.plot(x, y, label='sigmoid')
    plt.plot(x, dy, label='sigmoid gradient')
    plt.legend()
    plt.show()
    
plot_sigmoid()
```

```python
#5.Write in python a function to plot the ReLU activation function and its gradient using matplotlib.

import matplotlib.pyplot as plt
import numpy as np

def plot_relu():
    # Define the ReLU function and its gradient
    relu = lambda x: np.maximum(0, x)
    relu_grad = lambda x: np.where(x > 0, 1, 0)

    # Generate the input values
    x = np.linspace(-5, 5, 100)

    # Evaluate the ReLU function and its gradient at the input values
    y = relu(x)
    y_grad = relu_grad(x)

    # Plot the ReLU function and its gradient
    fig, ax = plt.subplots(2, 1, figsize=(6, 8))
    ax[0].plot(x, y)
    ax[0].set_title('ReLU')
    ax[1].plot(x, y_grad)
    ax[1].set_title('ReLU Gradient')
    plt.show()
    
plot_relu()
```

```python
#6.Write in python a function to plot the tanh activation function and its gradient using matplotlib.

import numpy as np
import matplotlib.pyplot as plt

def tanh_function(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x) ** 2

# Create a range of x-values to plot the functions
x_values = np.linspace(-5, 5, 1000)

# Compute the y-values for the tanh function and its derivative
y_tanh = tanh_function(x_values)
y_tanh_derivative = tanh_derivative(x_values)

# Plot the tanh function and its derivative
fig, ax = plt.subplots(2, 1, figsize=(8, 6))
ax[0].plot(x_values, y_tanh, label='tanh')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('tanh Activation Function')

ax[1].plot(x_values, y_tanh_derivative, label='tanh derivative')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_title('tanh Activation Function Gradient')

plt.tight_layout()
plt.show()
```

```python
#7.Write in python a function to plot the softmax activation function.

import matplotlib.pyplot as plt
import torch

def plot_softmax():
    x = torch.linspace(-5, 5, 100)
    y = torch.softmax(x, dim=0)

    plt.plot(x.numpy(), y.numpy())
    plt.title('Softmax Function')
    plt.xlabel('Input')
    plt.ylabel('Output')
    plt.show()
    
plot_softmax()
```

```python
#8.Write in python a function to plot the leaky ReLU activation function and its gradient using matplotlib.

import matplotlib.pyplot as plt
import torch

def leaky_relu(x, alpha=0.01):
    return torch.max(alpha*x, x)

def leaky_relu_grad(x, alpha=0.01):
    grad = torch.ones_like(x)
    grad[x < 0] = alpha
    return grad

def plot_leaky_relu():
    # Define input values
    x = torch.linspace(-5, 5, 100)

    # Compute output values and gradients
    y = leaky_relu(x)
    grad = leaky_relu_grad(x)

    # Plot function and gradient
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.plot(x.numpy(), y.numpy(), label='Leaky ReLU')
    ax.plot(x.numpy(), grad.numpy(), label='Gradient')
    ax.set_xlabel('Input')
    ax.set_ylabel('Output/Gradient')
    ax.set_title('Leaky ReLU Activation Function')
    ax.legend()
    plt.show()

# Call the function to plot the leaky ReLU activation function and its gradient
plot_leaky_relu()
```

```python
#9.Implement a binary classification model using the Perceptron class in PyTorch for the logic OR.
#Your task is to create a Perceptron instance and train it using a proper dataset and the binary cross-entropy loss with stochastic gradient descent optimizer.

import torch
import torch.nn as nn
import torch.optim as optim

inputs = torch.tensor([[0., 0.], [0., 1.], [1., 0.], [1., 1.]])
labels = torch.tensor([0., 1., 1., 1.]).unsqueeze(1)

# Define the Perceptron model
class Perceptron(nn.Module):
    def __init__(self):
        super(Perceptron, self).__init__()
        self.linear = nn.Linear(2, 1)
        self.activation = torch.nn.Sigmoid()
    
    def forward(self, x):
        x = self.linear(x)
        x = self.activation(x)
        return x

# Define the loss function and optimizer
perceptron = Perceptron()
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.SGD(perceptron.parameters(), lr=0.1)

for epoch in range(10000):
    outputs = perceptron(inputs)
    loss = criterion(outputs, labels)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

# Test the model
test_input = torch.tensor([1., 0.])
prediction = perceptron(test_input)
print("Prediction for [1, 0]:", prediction.item())

test_inputs = torch.tensor([[0., 0.], [0., 1.], [1., 0.], [1., 1.]])
true_outputs = torch.tensor([0., 1., 1., 1.]).unsqueeze(1)
predicted_outputs = perceptron(test_inputs)

test_predictions = (predicted_outputs >= 0.5).float()

correct_predictions = (test_predictions == true_outputs).sum().item()
accuracy = correct_predictions / len(true_outputs)

print("Accuracy:", accuracy)
```
